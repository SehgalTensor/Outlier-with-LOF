```python
#Question 3

import pandas as pd
import numpy as np
data = pd.read_csv('/EM_Outlier.csv')
user_id = data['user_id']
load_video = data['load_video']
pause_video = data['pause_video']
play_video = data['play_video']
seek_video = data['seek_video']
speed_change_video = data['speed_change_video']
stop_video = data['stop_video']


s = []
k = []
lrd2 = []
lrd3 = []
lof2 = []
lof3 = []
b1 = []
b2 = []
Mdis_k_2 = []
Edis_k_3 = []


for a in range(0,len(data)):
    k = [load_video[a],pause_video[a],play_video[a],seek_video[a],speed_change_video[a],stop_video[a]]
    s.append(k)
      
def M_dis(x,y):
    distance = 0
    for i in range(0,len(x)):
        distance = distance + abs(x[i]-y[i])
    return distance

def E_dis(x,y):
    distance = 0
    for i in range(0,len(x)):
        distance = distance + np.square(x[i]-y[i])
    return np.sqrt(distance)
    

def KNN_M(x,k):
    sort = []
    a1 = []
    for i in range(0,len(s)):
        a1 = M_dis(x,s[i])
        sort.append(a1)
    sort = sorted(sort)         
    return sort[k]

def KNN_E(x,k):
    sort1 = []
    a2 = []
    for i in range(0,len(s)):
        a2 = E_dis(x,s[i])
        sort1.append(a2)        
    sort1 = sorted(sort1)
    return sort1[k]


for i in range(0,len(s)):
    b1 = KNN_M(s[i],2)
    Mdis_k_2.append(b1)
    b2 = KNN_E(s[i],3)
    Edis_k_3.append(b2)
 
def N_M(x):    
    n1 = []
    for i in range(0,len(s)):
        if (M_dis(s[x],s[i])<=Mdis_k_2[x]):
            if (i != x):
                n1 = np.append(n1,i)
    return n1
               
def N_E(x):
    n2 = []
    for i in range(0,len(s)):
        if (E_dis(s[x],s[i])<=Edis_k_3[x]):
            if (i != x):
                n2 = np.append(n2,i)
    return n2            

for i in range(0,len(s)):
    a = 0 
    b = 0
    c = []
    d = []
    r1 = N_M(i)
    r2 = N_E(i)
    for j in range(0,len(r1)):
        t = int(r1[j])
        a = a + max(Mdis_k_2[t],M_dis(s[i],s[t]))
    c = (len(r1)/a)
    lrd2.append(c)
    for j in range(0,len(r2)):
        r = int(r2[j])
        b = b + max(Edis_k_3[r],E_dis(s[i],s[r]))
    d = (len(r2)/b)
    lrd3.append(d)
        
for i in range(0,len(s)):
    e = 0
    f = 0
    g = []
    h = []
    r1 = N_M(i)
    r2 = N_E(i)
    for j in range(0,len(r1)):
        l = int(r1[j])
        e = e + lrd2[l]
    g = (e/(len(r1)*lrd2[i]))
    lof2.append(g)
    for j in range(0,len(r2)):
        m = int(r2[j])
        f = f + lrd3[m]
    h = (f/(len(r2)*lrd3[i]))
    lof3.append(h)

lof2_5 = sorted(lof2,reverse = True)
lof3_5 = sorted(lof3,reverse = True)

print ("5 outliers and their LOF k=2:") 
for i in range(0,5):
    for j in range(0,len(s)):
        if (lof2_5[i] == lof2[j]):
            print(j,lof2_5[i])
         
print ("5 outliers and their LOF k=3:")
for i in range(0,5):
    for j in range(0,len(s)):
        if (lof3_5[i] == lof3[j]):
            print(j,lof3_5[i])   

```
