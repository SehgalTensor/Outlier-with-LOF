```python
import re
import numpy as np
import pandas as pd



df = pd.read_csv('/EM_Outlier.csv')


c1 = [1,1,1,1,1,1] 
c2 = [0,0,0,0,0,0]
n = 0


while n <= 50:
    lista = {}
    for i in range(0,(len(df.iloc[:,1]))):
        
        #Calculating weight 
        w11 = sum((df.iloc[i,1:]-c2)**2)/(sum((df.iloc[i,1:]-c2)**2) + sum((df.iloc[i,1:]-c1)**2))
        w12 = 1 - w11
        
        #adding the weight 
        lista[i] = [w11,w12]
    
        #Creating DataFrame 
        M = pd.DataFrame(list(lista.values())).transpose()
        
        
    #New centers
    c11 = []
    c22 = []
    for j in range(1,len(df.columns)):
        c11.append(sum(((M.iloc[0,:]**2)*df.iloc[:,j]))/sum((M.iloc[0,:]**2)))
        c22.append(sum(((M.iloc[1,:]**2)*df.iloc[:,j]))/sum((M.iloc[1,:]**2)))

    #SSE calculation
    sse1 = 0
    sse2 = 0
    if n<3 or n==29:
           
        for i in range(0,(len(df.iloc[:,1]))):
            dist1 = sum((df.iloc[i,1:]-c11)**2)
            dist2 = sum((df.iloc[i,1:]-c22)**2)
            sse1 +=  (dist1) * lista[i][0]  
            sse2 += (dist2) * lista[i][1] 
            
        
    
    d1 = sum(np.abs(np.array(c1) - np.array(c11)))
    d2 = sum(np.abs(np.array(c2) - np.array(c22)))
    n += 1
    if (n < 3): #Display the updated centers and SSE only for two iterations
        print('Iteration: '),
        print(n)
        print('')
        print('Updated Centers: ')
        print(c11); print(c22)
        print('')
        print('SSE for two clusters: ')
        print(sse1)
        print("|")
        print(sse2)
        
        
    if (d1 < 0.001) & (d2 < 0.001): #Terminate if sum of L1 distances are smaller than 0.001
        print('Iteration: '),
        print(n)
        print('')
        print('Updated Centers: ')
        print(c11); print(c22)
        break
    else:
        c1 = c11
        c2 = c22
    
```
